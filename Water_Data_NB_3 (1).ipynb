```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Reading the file
df=pd.read_csv('Results_MADE.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Classifiying in 5 classes
0-25: Excellent: 0
25-50: Very Good: 1
50-75: Good: 2
75-100: Bad: 3
100-above: Not Drinkable: 4   

```python
df['Class']=0
```

```python
df['Class'][df['WQI']>=0]=0
df['Class'][df['WQI']>25]=1
df['Class'][df['WQI']>50]=2
df['Class'][df['WQI']>75]=3
df['Class'][df['WQI']>100]=4
```

```python
df.head(10)
```

```python
df['Class'].value_counts()
```

```python
df.drop(['WQI'],axis=1,inplace=True)
df.head()
```

```python
#df.to_csv('Data_Classified.csv')
```

```python
plt.pie([24,81,85,44,61],labels=['0','1','2','3','4'],autopct='%2.2f%%')
plt.title('Water Classified by WQI')
plt.show()
```

```python
#Creating X & y variables
X=df[['Temperature','Dissolved Oxygen','pH','Bio-Chemical Oxygen Demand (mg/L)','Faecal Streptococci (MPN/ 100 mL)','Nitrate (mg/ L)','Faecal Coliform (MPN/ 100 mL)','Total Coliform (MPN/ 100 mL)','Conductivity (mho/ Cm)']]
y=df['Class']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10,test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_train=pd.DataFrame(X_train)
```

```python
# import pickle
# pickle.dump(sc,open('scaler.pkl','wb'))
```

```python
X_train
```

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
y_pred=model_svc.predict(sc.transform(X_test))
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
kernels=['linear','poly','rbf','sigmoid']
c=[0.01,0.1,1.0,10.0,100.0]
param_grid={'C':c,
           'kernel': kernels}
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
grid_predictions=grid.predict(sc.transform(X_test))
```

```python
grid_predictions
```

```python
print(classification_report(y_test, grid_predictions))
```

TRYING WITH ANN

```python
import tensorflow as tf
```

```python
#ANN
ann=tf.keras.models.Sequential()
ann
```

```python
#Adding layer (Input)
ann.add(tf.keras.layers.Dense(units=9,activation='relu'))
```

```python
#Adding layer (1st hidden layer)
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
#Adding layer (Output)
ann.add(tf.keras.layers.Dense(units=4,activation='softmax'))
```

```python
#Stochastic Gradient Descent
ann.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
y_train
```

```python
y_train_df=pd.DataFrame(y_train.values,columns=['WQI'])
y_train_df.head(20)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('tnf3',OneHotEncoder(sparse=False,drop='first'),['WQI'])],remainder='passthrough')
```

```python
y_train_df=np.array(ct.fit_transform(y_train_df))
```

```python
y_train_df
```

```python
#Training the model
ann.fit(X_train,y_train_df,batch_size=32,epochs=100)
```

```python
#Predictions
y_pred=ann.predict(sc.transform(X_test))#Will return a decimal value
y_pred
```

```python
y_test
```
